```python
#Exercise 1
import math
class point: 
    def __init__(self,X,Y):
        self.X=X
        self.Y=Y

    def __str__(self):
        return"I am a point in the two dimentional space X Y  "
        
    def distance(self,other):
            dis=math.sqrt((self.X - other.X )**2 + (self.Y - other.Y)**2)
            return (dis)
        


point1=point(3,5)
point2=point(4,8)
print(point2.distance(point1))

```
